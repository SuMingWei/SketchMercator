```python
from python_lib.pkl_saver import PklSaver
out_saver = PklSaver(".", "data.pkl")
data_dict = out_saver.load()
```

```python
plot_data_dict = {}

for workload_name, v1 in data_dict.items():
    print(workload_name)
    plot_data_dict[workload_name] = []
    for i in range(14):
        plot_data_dict[workload_name].append([])

    
    distribution_list = {}
    for bf, v2 in v1.items():
        distribution_list[bf] = []
#         print(bf)
        for num_sketches, v3 in v2.items():
#             print(num_sketches)
            distribution = [0] * 14
            for elem in v3:
                distribution[elem] += 1
#             print(distribution)
            distribution_list[bf].append(distribution)
    for before_list, after_list  in zip(distribution_list['before'], distribution_list['after']):
        for i in range(1, 14):
            plot_data_dict[workload_name][i].append(before_list[i])
            plot_data_dict[workload_name][i].append(after_list[i])

#         print(before_list)
#         print(after_list)
#         print()


```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib


matplotlib.rcParams['hatch.linewidth'] = 0.1
matplotlib.rcParams['text.usetex'] = True
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42


for workload_name, v in plot_data_dict.items():
    if "1" in workload_name:
        plt.figure(figsize=(8, 3.5))
    else:
        plt.figure(figsize=(8, 2.7))
    
    x = [i for i in range(1, 21)]
    x_list = []
    for i in range(1, 11):
        x_list.append(i-0.2)
        x_list.append(i+0.2)
    x = np.array(x_list)
    y_sum = np.array([0] * 20)
    for i in range(1, 14):
        color = "C%d" % i
        if i == 13:
            label = "X"
        else:
            label = str(i)
        y = np.array(plot_data_dict[workload_name][i])
        plt.bar(x, y, width=0.3, bottom=y_sum, alpha=0.4, color = color, label=label)
        for xpos, ypos, y_elem in zip(x, y_sum+y/2, y):
            if y_elem != 0:
                plt.text(xpos, ypos, label, ha="center", va="center", fontsize=18)

        y_sum += y
    
    if "1" in workload_name:
        pingpong = True
        for xpos, ypos in zip(x, y_sum):
            if pingpong:
                label = "be"
                pingpong = False
            else:
                label = "af"
                pingpong = True

            plt.text(xpos, ypos+0.6, label, ha="center", va="center", fontsize=18)

    ax = plt.gca()
    from matplotlib.ticker import MultipleLocator
    ax.yaxis.set_major_locator(MultipleLocator(2))
    plt.yticks(fontsize=20)
    title_font_size = 22
    if "1" in workload_name:
        plt.title("Ensemble Type 1. Same sketch algorithm", fontsize=title_font_size, y=1.3)

    if "2" in workload_name:
        plt.title("Ensemble Type 2. Same flowkey", fontsize=title_font_size)

    if "3" in workload_name:
        plt.title("Ensemble Type 3. Same flowsize", fontsize=title_font_size)

#     if "4" in workload_name:
#         plt.title("Type 4. same epoch", fontsize=20)

    if "5" in workload_name:
        plt.title("Ensemble Type 4. Random", fontsize=title_font_size)
    if "1" in workload_name:
        ax.set_xlabel("number of sketch instances", fontsize=20)
    ax.set_ylabel('Fraction of \n Ten Ensembles', fontsize=20)
    plt.grid(color='gray', linestyle='--', linewidth=1, axis='y')
#     plt.legend()
#     plt.rcParams['xtick.labelbottom'] = False
#     plt.rcParams['xtick.bottom'] = False
#     plt.gca().xaxis.set_major_locator(plt.NullLocator())
#     plt.tick_params(bottom = False)
#     ax.tick_params(axis=u'both', which=u'both',length=0)
#     ax.spines['bottom'].set_visible(False)
#     plt.tick_params(top='off', bottom='off', left='off', right='off', labelleft='off', labelbottom='on')
#     ax.tick_params(tick1On=False)

    new_x = [i for i in range(1, 11, 1)]
    new_x_label = [i for i in range(2, 21, 2)]

#     print(new_x)
#     print(new_x_label)

    ax.set_xticks(new_x)
    ax.set_xticklabels(new_x_label, fontsize=20)


    plt.tight_layout()
#     latex_path = "/Users/hnamkung/hun-latex/Hun-multi-dim-sketches/nsdi23/figures/6_evaluation"
    latex_path = "."
    file_name = '%s.pdf' % workload_name.replace("_", "")
    import os
    full_path = os.path.join(latex_path, file_name)
    plt.savefig(full_path, bbox_inches='tight', pad_inches=0.02)
    plt.show()
#     break



    
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = ["a","b","c","d"]
y1 = np.array([3,8,6,4])
y2 = np.array([10,2,4,3])
y3 = np.array([5,6,2,5])

# memo of sample number
snum = y1+y2+y3

# normalization
y1 = y1/snum*100.
y2 = y2/snum*100.
y3 = y3/snum*100.

print(y1)
print(y2)
print(y3)

```

```python
plt.figure(figsize=(10, 3))
x = [1, 2, 9, 10]
# stack bars
plt.bar(x, y1, width=0.3, label='y1')
plt.bar(x, y2, width=0.3, bottom=y1,label='y2')
plt.bar(x, y3, width=0.3, bottom=y1+y2,label='y3')

# add text annotation corresponding to the percentage of each data.
for xpos, ypos, yval in zip(x, y1/2, y1):
    plt.text(xpos, ypos, "%.1f"%yval, ha="center", va="center")

for xpos, ypos, yval in zip(x, y1+y2/2, y2):
    plt.text(xpos, ypos, "%.1f"%yval, ha="center", va="center")

for xpos, ypos, yval in zip(x, y1+y2+y3/2, y3):
    plt.text(xpos, ypos, "%.1f"%yval, ha="center", va="center")


for xpos, ypos, yval in zip(x, y1+y2+y3, snum):
    plt.text(xpos, ypos, "N=%d"%yval, ha="center", va="bottom")

ax = plt.gca()
# ax.set_xlabel(["a", "b", "c", "d"], fontsize=20)

plt.ylim(0,110)
ax.set_xticks([2, 4, 6, 8])
ax.set_xticklabels([2, 4, 6, 8], fontsize=10)


plt.legend(bbox_to_anchor=(1.01,0.5), loc='center left')
plt.show()
# plt.savefig('normalized_stacked_barplot_with_number.png', bbox_inches='tight', pad_inches=0.02)
```
