```python
plot_data_dict = {}

for workload_name, v1 in data_dict.items():
    print(workload_name)
    plot_data_dict[workload_name] = []
    for i in range(14):
        plot_data_dict[workload_name].append([])

    
    distribution_list = {}
    for bf, v2 in v1.items():
        distribution_list[bf] = []
#         print(bf)
        for num_sketches, v3 in v2.items():
#             print(num_sketches)
            distribution = [0] * 14
            for elem in v3:
                distribution[elem] += 1
#             print(distribution)
            distribution_list[bf].append(distribution)
    for before_list, after_list  in zip(distribution_list['before'], distribution_list['after']):
        for i in range(1, 14):
            plot_data_dict[workload_name][i].append(before_list[i])
            plot_data_dict[workload_name][i].append(after_list[i])

#         print(before_list)
#         print(after_list)
#         print()


```
